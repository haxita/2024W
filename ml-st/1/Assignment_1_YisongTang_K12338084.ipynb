<h2 style="color:rgb(210,90,80)">Task 1: Risk Calulation (10 Points)</h2>

Assume the simple situation of a binary classification task, i.e. $y = \pm 1$, and a 1-dimensional feature $x$, i.e. $d=1$. Moreover, the feature is discrete (categorical) and can only have 3 possible values $x=1,2,3$.

Data is drawn from the following distribution:

\begin{align}
p(x=1,y=+1) &= 0.1\\
p(x=1,y=-1) &= 0.2\\
p(x=2,y=+1) &= 0.2\\
p(x=2,y=-1) &= 0.1\\
p(x=3,y=+1) &= 0.3\\
p(x=3,y=-1) &= 0.1\\
\end{align}

Compute the risk for the 0-1-loss for these two classifiers:

(1) The Bayes-optimal classifier $g_\textrm{opt}$.<br>
(2) A classifier $g_\textrm{dom}$ which always outputs $+1$, i.e. the dominant label, independent of $x$.

YOUR ANSWER HERE

Q1.\
Step 1: Compute Marginal Probabilities $p(x)$

The marginal probability $p(x)$ is obtained by summing over all possible labels  $y$:

\begin{align*}
p(x=1) &= p(x=1, y=+1) + p(x=1, y=-1) = 0.1 + 0.2 = 0.3, \\
p(x=2) &= p(x=2, y=+1) + p(x=2, y=-1) = 0.2 + 0.1 = 0.3, \\
p(x=3) &= p(x=3, y=+1) + p(x=3, y=-1) = 0.3 + 0.1 = 0.4.
\end{align*}


Step 2: Compute Conditional Probabilities $p(y | x)$

Using Bayes' theorem, the conditional probabilities are:

\begin{align*}
p(y=+1 \mid x=1) &= \frac{p(x=1, y=+1)}{p(x=1)} = \frac{0.1}{0.3} = \frac{1}{3}, \\
p(y=-1 \mid x=1) &= \frac{p(x=1, y=-1)}{p(x=1)} = \frac{0.2}{0.3} = \frac{2}{3}, \\
p(y=+1 \mid x=2) &= \frac{p(x=2, y=+1)}{p(x=2)} = \frac{0.2}{0.3} = \frac{2}{3}, \\
p(y=-1 \mid x=2) &= \frac{p(x=2, y=-1)}{p(x=2)} = \frac{0.1}{0.3} = \frac{1}{3}, \\
p(y=+1 \mid x=3) &= \frac{p(x=3, y=+1)}{p(x=3)} = \frac{0.3}{0.4} = \frac{3}{4}, \\
p(y=-1 \mid x=3) &= \frac{p(x=3, y=-1)}{p(x=3)} = \frac{0.1}{0.4} = \frac{1}{4}.
\end{align*}

Bayes-Optimal Classifier $g_{\text{opt}}$

The Bayes-optimal classifier selects the label with the highest posterior probability $p(y | x)$ for each $x$:

\begin{align*}
x = 1 &: \text{Choose } y = -1 \quad \left(\frac{2}{3} > \frac{1}{3}\right), \\
x = 2 &: \text{Choose } y = +1 \quad \left(\frac{2}{3} > \frac{1}{3}\right), \\
x = 3 &: \text{Choose } y = +1 \quad \left(\frac{3}{4} > \frac{1}{4}\right).
\end{align*}


Risk Calculation:

The risk under 0-1 loss is the probability of misclassification:

\begin{align*}
\text{At } x=1 &: \text{Misclassify } y=+1 \quad \Rightarrow \quad p(x=1, y=+1) = 0.1, \\
\text{At } x=2 &: \text{Misclassify } y=-1 \quad \Rightarrow \quad p(x=2, y=-1) = 0.1, \\
\text{At } x=3 &: \text{Misclassify } y=-1 \quad \Rightarrow \quad p(x=3, y=-1) = 0.1.
\end{align*}

Total Risk $= 0.1 + 0.1 + 0.1 = 0.3$


Q2. Dominant Classifier $g_{\text{dom}}$

This classifier always predicts $y = +1$ regardless of $x$

Risk Calculation:

The risk is the probability that the true label $y$ is not $+1$:

$Risk = p(y=-1) = p(x=1, y=-1) + p(x=2, y=-1) + p(x=3, y=-1) = 0.2 + 0.1 + 0.1 = 0.4$

<h3 style="color:rgb(210,90,80)">Code 3.1 (5 Points):</h3>

```python
def train_kNN(X_train: np.ndarray, y_train: np.ndarray, k: int) -> neighbors.KNeighborsClassifier: 
    """Function that fits a kNN to given data.
    
    Parameters
    ----------
    X_train : np array
        Training data.
    y_train : np array
        Training data labels.
    k : int 
        How many neighbors to consider.

    Returns
    -------
    classifier : neighbors.KNeighborsClassifier
        Classifier that was fitted to the training data.
    """
    # YOUR CODE HERE
    # Initialize the k-NN classifier with the specified number of neighbors
    classifier = neighbors.KNeighborsClassifier(n_neighbors=k)
    
    # Fit the classifier to the training data
    classifier.fit(X_train, y_train)
    
    return classifier
```

```python
def eval_kNN(classifier: neighbors.KNeighborsClassifier, X_eval: np.ndarray) -> np.ndarray: 
    """Function that returns predictions for some input data.

    Parameters
    ----------
    classifier : neighbors.KNeighborsClassifier
        Trained kNN classifier.
    X_eval : np.ndarray
        Data used for predicting the labels.

    Returns
    -------
    predictions : np.ndarray
        Predicted labels.
    """
    # YOUR CODE HERE
    # Use the classifier to predict labels for the evaluation data
    predictions = classifier.predict(X_eval)
    
    return predictions
```

<h3 style="color:rgb(210,90,80)">Code 3.3 (5 Points):</h3>

```python
def mean_zero_one_loss(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    """Function that calculates the mean zero-one loss for given true and predicted labels.

    Parameters
    ----------
    y_true : np.ndarray
        True labels.
    y_pred : np.ndarray
        Tredicted labels.

    Returns
    -------
    loss : float
        Mean zero-one loss.
    """
    # YOUR CODE HERE
    # Calculate the number of misclassifications
    misclassifications = np.sum(y_true != y_pred)
    
    # Compute the mean zero-one loss
    loss = misclassifications / len(y_true)
    
    return loss
```

<h3 style="color:rgb(210,90,80)">Code 3.4 (5 Points):</h3>

```python
def run_kNN(X: np.ndarray, y: np.ndarray, train_kNN: Callable, eval_kNN: Callable, mean_zero_one_loss: Callable, nf: int, k: int) -> float:
    """Function that combines all previous functions using Cross-Validation.
    
    Parameters
    ----------
    X : np.ndarray
        Training data.
    y : np.ndarray
        Training labels.
    train : Callable
        Function used for training the clf.
    evaluate : Callable
        Function used for evaluating the clf.
    loss_fn : Callable
        Function used for calculating the loss.
    nf : int
        Number of folds for CV
    k : int
        k for kNN algorithm.

    Returns
    -------
    mean_error : float
        Mean error over all folds.
    """
    # YOUR CODE HERE
    # Initialize K-Fold cross-validation
    kf = KFold(n_splits=nf, shuffle=True, random_state=42)
    
    # List to store loss for each fold
    loss_list = []
    
    # Iterate over each fold
    for train_index, test_index in kf.split(X):
        # Split data into training and testing for this fold
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        # Train the k-NN classifier
        classifier = train_kNN(X_train, y_train, k)
        
        # Predict on the test set
        y_pred = eval_kNN(classifier, X_test)
        
        # Calculate the loss for this fold
        loss = mean_zero_one_loss(y_test, y_pred)
        
        # Append the loss to the list
        loss_list.append(loss)
    
    # Compute the mean error over all folds
    mean_error = np.mean(loss_list)
    
    return mean_error
```

```python
def generate_flip_vector(n: int) -> np.ndarray:
    """Function that produces a flip vector consisting of -1's and 1's.

    Parameters
    ----------
    n : integer
        The length of the vector that should be returned.

    Returns
    -------
    flip_vector : np.ndarray
        The vector that indicates what labels will be flipped
    """
    # YOUR CODE HERE
    # Calculate the number of -1s
    num_neg = n // 5  # floor division
    
    # Calculate the number of +1s
    num_pos = n - num_neg
    
    # Create an array with num_pos +1's and num_neg -1's
    flip_vector = np.array([1] * num_pos + [-1] * num_neg)
    
    # Shuffle the flip_vector in-place
    np.random.shuffle(flip_vector)
    
    return flip_vector
```

```python
def plot_subplots(X: np.ndarray, y: np.ndarray, y_fl: np.ndarray, m: int, error_holder: list[float], error_holder_flipped: list[float]):
    """The function creating the plots as described above.

    The Parameters are called the same as in the notebook.
    Do not use anything besides them.

    Parameters
    ----------
    X : np.ndarray
        The data to plot.
    y : np.ndarray
        The original labels of the data.
    y_fl : np.ndarray
        The flipped labels of the data.
    m : range
        The range of the error_holder lists.
    error_holder : list[float]
        The list containing the original errors.
    error_holder_flipped : list[float]
        The list containing the errors on the fipped dataset labels.  
    """
    # YOUR CODE HERE
    # Create a figure with 2x2 subplots
    fig, axs = plt.subplots(2, 2, figsize=(16, 12))
    
    # Top Left: Original Dataset
    ax1 = axs[0, 0]
    colors = {1: 'blue', -1: 'red'}
    markers = {1: 'o', -1: 's'}
    for label in np.unique(y):
        idx = y == label
        ax1.scatter(
            X[idx, 0],
            X[idx, 1],
            c=colors[label],
            marker=markers[label],
            label=f'y={int(label)}',
            edgecolor='k',
            alpha=0.7
        )
    ax1.set_xlabel('$x_1$', fontsize=12)
    ax1.set_ylabel('$x_2$', fontsize=12)
    ax1.set_title('Original Dataset', fontsize=14)
    ax1.legend(title='Classes')
    
    # Top Right: Flipped Dataset
    ax2 = axs[0, 1]
    for label in np.unique(y_fl):
        idx = y_fl == label
        ax2.scatter(
            X[idx, 0],
            X[idx, 1],
            c=colors[label],
            marker=markers[label],
            label=f'y={int(label)}',
            edgecolor='k',
            alpha=0.7
        )
    ax2.set_xlabel('$x_1$', fontsize=12)
    ax2.set_ylabel('$x_2$', fontsize=12)
    ax2.set_title('Flipped Dataset', fontsize=14)
    ax2.legend(title='Classes')
    
    # Bottom Left: Mean Error vs k for Original Dataset
    ax3 = axs[1, 0]
    ax3.plot(m, error_holder, marker='o', linestyle='-', color='b')
    ax3.set_xlabel('k (Number of Neighbors)', fontsize=12)
    ax3.set_ylabel('Mean Zero-One Loss', fontsize=12)
    ax3.set_title('Mean Zero-One Loss vs k (Original)', fontsize=14)
    ax3.grid(True, linestyle='--', alpha=0.7)
    
    # Bottom Right: Mean Error vs k for Flipped Dataset
    ax4 = axs[1, 1]
    ax4.plot(m, error_holder_flipped, marker='o', linestyle='-', color='r')
    ax4.set_xlabel('k (Number of Neighbors)', fontsize=12)
    ax4.set_ylabel('Mean Zero-One Loss', fontsize=12)
    ax4.set_title('Mean Zero-One Loss vs k (Flipped)', fontsize=14)
    ax4.grid(True, linestyle='--', alpha=0.7)
    
    # Adjust layout
    plt.tight_layout()
    
    # Display the plot
    plt.show()
    
    # Optionally, return the figure object
    return fig
```

As before, plot the error versus $k$ for 10-folds CV with 1, 2, 3, 4 incrementally added features. (4 plots) <br>

Additionally, for the particular choice $k=11$, plot the mean error versus $f$ with $f = 2,3,4,5,6$ being the number of features. Thus, the first data point in this plot (where $f=2$) shows the error for the original feature matrix X without extra dimensions. The second data point in this plot (where $f=3$) shows the error for the feature matrix X with one additional dimension, and so on.<br>

<h3 style="color:rgb(210,90,80)">Code 5.1 (10 Points):</h3>

```python
def add_features(X: np.ndarray) -> np.ndarray:
    """Function that adds uniformly distributed features to dataset.

    Parameters
    ----------
    X : np.ndarray
        Dataset to add random feature

    Returns
    -------
    X_new : np.ndarray
        Dataset enhanced with 4 uniformly distributed features.
    """
    
    # YOUR CODE HERE
     # Generate 4 uniformly distributed features between 0 and 1
    extra_features = np.random.uniform(low=0.0, high=1.0, size=(X.shape[0], 4))
    
    # Concatenate the new features to the original X
    X_new = np.hstack((X, extra_features))
    
    return X_new
```

<h3 style="color:rgb(210,90,80)">Plot 5.2 (10 Points):</h3>

```python
def plot_error_vs_k_extra_dims(X_new: np.ndarray, y: np.ndarray, m: range):
    """Function that implements the plot from Task5.

    This function should create 2! plots.
    The first plot to visualize error vs k with 10 folds and <X> extra dimensions.
    The second plot to visualize the error versus dimension with k=11
    
    Parameters
    ----------
    X_new : np.ndarray
        The new dataset with the added random features.
    y : np.ndarray
        The labels of the dataset.
    m : range
        The range of the error_holder lists for the new errors.
    """
    # YOUR CODE HERE
    # Define the range of f values (number of features)
    f_values = range(2, 7)  # f = 2,3,4,5,6

    # Initialize a dictionary to hold error vs k for each f
    error_vs_k = {}

    # Initialize a list to hold error vs f for k=11
    error_vs_f = []

    # Iterate over f values
    for f in f_values:
        # Select the first f features
        X_f = X_new[:, :f]
        
        # Initialize list to hold errors for all k in m
        error_holder = []
        
        # Iterate over k
        for k in m:
            # Compute mean error via run_kNN
            mean_error = run_kNN(X_f, y, train_kNN, eval_kNN, mean_zero_one_loss, nf=10, k=k)
            error_holder.append(mean_error)
        
        # Store in the dictionary
        error_vs_k[f] = error_holder
        
        # Now, for k=11, get the mean error
        mean_error_k11 = run_kNN(X_f, y, train_kNN, eval_kNN, mean_zero_one_loss, nf=10, k=11)
        error_vs_f.append(mean_error_k11)
    
    # Now, create the plots
    # First plot: error vs k for each f
    plt.figure(figsize=(14, 6))
    
    # Subplot 1: Mean Zero-One Loss vs k for different numbers of features
    plt.subplot(1, 2, 1)
    for f in f_values:
        plt.plot(m, error_vs_k[f], label=f'f={f}')
    plt.xlabel('k (Number of Neighbors)', fontsize=12)
    plt.ylabel('Mean Zero-One Loss', fontsize=12)
    plt.title('Mean Zero-One Loss vs k for Different Numbers of Features', fontsize=14)
    plt.legend(title='Number of Features')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Subplot 2: Mean Zero-One Loss vs number of features for k=11
    plt.subplot(1, 2, 2)
    plt.plot(f_values, error_vs_f, marker='o', linestyle='-', color='g')
    plt.xlabel('Number of Features (f)', fontsize=12)
    plt.ylabel('Mean Zero-One Loss', fontsize=12)
    plt.title('Mean Zero-One Loss vs Number of Features for k=11', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(f_values)
    
    # Adjust layout and display the plots
    plt.tight_layout()
    plt.show()
```

```python
# Nothing to do here - just run this cell!
plot_error_vs_k_extra_dims(X_new, y, m)
```
