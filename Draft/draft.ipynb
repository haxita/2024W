## Part 2: Express \( w = \left( \cos \dfrac{\pi}{6} - i\sin \dfrac{\pi}{6} \right)^7 \) in Trigonometric Form

### Note on the Expression:
$ \( \cos \dfrac{\pi}{6} - i\sin \dfrac{\pi}{6} \) $ is equivalent to $\( \cos\left( -\dfrac{\pi}{6} \right) + i\sin\left( -\dfrac{\pi}{6} \right) \) $ due to the identities:
$ \( \cos(-\theta) = \cos \theta \)$
$ \( \sin(-\theta) = -\sin \theta \)$

### Step 1: Recognize the Base in Trigonometric Form

Let:

$$
w_0 = \cos\left( -\dfrac{\pi}{6} \right) + i\sin\left( -\dfrac{\pi}{6} \right)
$$

### Step 2: Apply De Moivre’s Theorem

$$
w = \left( w_0 \right)^7 = \cos\left( 7 \cdot \left( -\dfrac{\pi}{6} \right) \right) + i\sin\left( 7 \cdot \left( -\dfrac{\pi}{6} \right) \right)
$$

Compute the exponentiated angle:

$$
7 \left( -\dfrac{\pi}{6} \right) = -\dfrac{7\pi}{6}
$$

### Step 3: Adjust the Angle to the Principal Value

Add \( 2\pi \) to bring the angle into the interval \([0, 2\pi)\):

$$
-\dfrac{7\pi}{6} + 2\pi = -\dfrac{7\pi}{6} + \dfrac{12\pi}{6} = \dfrac{5\pi}{6}
$$

### Step 4: Write \( w \) in Trigonometric Form

$$
w = \cos\left( \dfrac{5\pi}{6} \right) + i\sin\left( \dfrac{5\pi}{6} \right)
$$

### Verification of Modulus:

Since the modulus of \( w_0 \) is \( 1 \):

$$
|w| = |w_0|^7 = 1^7 = 1
$$

Therefore, \( w \) has modulus \( 1 \).
